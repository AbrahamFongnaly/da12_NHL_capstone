```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
# Define the path to your CSV files
path = 'C:\DA12\capstone\da12_NHL_capstone\data'

# Get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(path, "*.csv"))

# Create an empty list to store DataFrames
dfs = []

# Loop over the list of CSV files and read each file
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

# Combine all DataFrames into one (optional)
nhl_goalie_df = pd.concat(dfs, ignore_index=True)

# Display the first few rows of the combined DataFrame
print(f"Loaded {len(dfs)} CSV files.")
nhl_goalie_df
```

```python
#clean column (xPlay Stopped and play Stopped) names to make easier for future coding
nhl_goalie_df = nhl_goalie_df.rename(columns = {'xPlay Stopped':'xPlayStopped','play Stopped':'playStopped'})
```

```python
all_situation_goalie['save_percentage'] = ((all_situation_goalie['ongoal'] - all_situation_goalie['goals']) / all_situation_goalie['ongoal']) * 100
#create new column 'save percentage' by subtracting scored goals from all shots and dividing result from all shots on goal
#multiply by 100 
```

```python
all_situation_goalie['save_percentage'] = round(all_situation_goalie['save_percentage'],2)
```
